## V Ellipses de MacAdam

```python
colPlt.plot_ellipses_MacAdam1942_in_chromaticity_diagram_CIE1931()
```

Les ellipses de MacAdam représentent les zone où l'oeil humain ne peut pas distinguer les nuances de couleurs. On constate que les ellipses sont plus grandes dans la zone verte, la couleur la plus lumineuse du spectre. Le standard CIE 1931 n'est donc pas de plus optimisé pour permettre à l'homme de distinguer les nuances.

```python
colPlt.plot_ellipses_MacAdam1942_in_chromaticity_diagram_CIE1960UCS()
```

On constate sur le standard CIE 1960 une reduction de la taille des ellipses de MacAdam. Mais globalement, leurs emplacements sont les mêmes que sur le diagramme du standard CIE 1931.
Ce standard est donc plus optimisé mais pas optimal.

```python
colPlt.plot_ellipses_MacAdam1942_in_chromaticity_diagram_CIE1976UCS()
```

On constate sur le standard CIE 1976 que les ellipses de MacAdam sont strictement les mêmes que sur le standard CIE 1960.

## VI Espaces de couleur

```python
# On affiche par exemple les informations de l'espace Gamut 3
print(rgb['S-Gamut3'])
```

Un espace de couleur est un standard qui permet un calibrage pour afficher une image sur un moniteur ou l'imprimer selon un certain seuil de qualité.

Dans chaque espace de couleurs sont stockées les coordonnées des principales couleurs de l'espace : les couleurs, le point le plus blanc, le NPM et le NPM inverse, le NPM dérivé et le NPM dérivé inverse.

Les données sont stockées sous forme de tableaux, exploitables facilement avec Python.
